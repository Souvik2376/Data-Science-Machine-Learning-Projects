# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Set

```python
#dataset impliment and read using pandas and there is no header in the top coloumn
#thats why we use header =None
df = pd.read_csv("sonar data.csv" , header=None)
```

M ---> Mine
R ---> Rock

```python
#grouping
df.groupby(60).mean()
```

```python
# Separating data and labels
x = df.drop(columns=60, axis=1)
y =df[60]
```

```python
print(x)
print(y)
```

Training & test data

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = 0.1, stratify=y,random_state=1)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

# Model = Logistrick regression

```python
model = LogisticRegression()
```

```python
print(x_train,y_train)
```

```python
#trainning the logistic regression data
model.fit(x_train, y_train)
```

Model eveluation

```python
#accuracy on the trainning data
x_train_prediction = model.predict(x_train)
trainning_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on trainning data : ' , trainning_data_accuracy)
```

```python
#accuracy on the test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on test data : ' , test_data_accuracy)
```
